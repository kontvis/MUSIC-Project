Next step: loop over each year and pull all the songs on the Billboard Top 100 from each year, add a column for the year.

##### Notes on the URL structure of the Billboard Top 100
The Billboard Top 100 has the top 100 songs for each year back to 1940. 
The URLs follow this standard: http://billboardtop100of.com/YYYY-2/ <br>

<b>Except</b> for 1940 which does nto follow the stucture: http://billboardtop100of.com/336-2/ <br>
Also, the year 1940 does not have the songs listed in a table. 
Therefore, for simplicity, I will only pull the songs from 1941 - current year.

```python
from bs4 import BeautifulSoup as soup
from urllib.request import urlopen as uReq
import pandas as pd
```

```python
#Open web client connection and grab the webpage, save it as a variable.
myurl = "http://billboardtop100of.com/2000-2/"
uClient = uReq(myurl)
page_html = uClient.read()
```

```python
#Close the web client connection.
uClient.close()
```

```python
#This command parses the html from the entire page.
page_soup = soup(page_html, "html.parser")
```

```python
#See the page header
table = page_soup.find_all('table')[0] # Grab the first table
```

```python
#put the contents of the table into a list
rows = table.find_all('tr')

data = []
for row in rows:
    cols = []
    for col in row.find_all('td'):
        t = col.text.strip()
        if not t:
            for img in row.find_all('img'):
                t = img.attrs['src']

        cols.append(t)
    data.append(cols)
```

```python
#create column headers, and a datafram of the labels and data
labels = ['rank', 'artist', 'song_title']
df = pd.DataFrame.from_records(data, columns=labels)
```
